#### Group companies by first letter(lower) and take the mean incidents_85_99 and make it a dataframe again 

```python
import pandas as pd
```

```python
df_new = df.copy()
```

```python
df_new["airline"] = df_new["airline"].astype(str).str[0].str.lower()
```

```python
df_new.head()
```

```python
df_new = df_new.rename(columns={"airline":"airline_fist_letter"})
```

```python
df_new.groupby("airline_fist_letter", axis=0).mean()["incidents_85_99"].to_frame()
```
